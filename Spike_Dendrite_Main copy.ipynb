```python
import deep_learning as dl
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
# silence 80% of feedback weights
dl.use_sparse_feedback = True
```

```python
# set training parameters

f_etas = ( 0.01,  0.01, 0.21, .21)
b_etas = ( 0.01, 0.01,0.12,0.12)
n_epochs = 60
n_training_examples = 60000
```

```python
# create the network -- this will also load the MNIST dataset files
net = dl.Network(n=(10))
```

```python
# train the network for x epochs
net.train(f_etas, b_etas, n_epochs, n_training_examples, save_simulation=True, simulations_folder="Simulations", folder_name="Example Simulation/test01")
```
